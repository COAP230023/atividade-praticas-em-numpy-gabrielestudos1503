### 31. Normalização min–max por coluna  _(médio)_

### 37. k-means 2D (NumPy puro)  _(difícil)_

Implemente k-means com K=3 em pontos 2D: inicialize centróides aleatórios, faça 10 iterações de (1) atribuir pontos ao centróide mais próximo e (2) recomputar centróides.

```python
import numpy as np

rng = np.random.default_rng(42)

X = rng.normal(size=(100, 2))
K = 3

indices = rng.choice(X.shape[0], size=K, replace=False)
centroids = X[indices].copy()

for _ in range(10):
    diff = X[:, None, :] - centroids[None, :, :]
    dists = np.linalg.norm(diff, axis=2)
    labels = np.argmin(dists, axis=1)

    for k in range(K):
        mask = labels == k
        if np.any(mask):
            centroids[k] = X[mask].mean(axis=0)

print("Centróides finais:\n", centroids)
print("Rótulos dos pontos:\n", labels)
```

### 38. PCA básica (autovetores)  _(difícil)_

Implemente PCA 2D→1D: centralize `X` (N×2), calcule covariância, autovetores, projete em 1D e reconstrua aproximado.

```python
import numpy as np

rng = np.random.default_rng(42)

X = rng.normal(size=(100, 2))

mean = X.mean(axis=0)
Xc = X - mean

C = np.cov(Xc, rowvar=False)
vals, vecs = np.linalg.eigh(C)
w = vecs[:, np.argmax(vals)]
z = Xc @ w
X_recon = np.outer(z, w) + mean

print("Componente principal (w):\n", w)
print("\nProjeções 1D (z):\n", z[:10])
print("\nPrimeiros pontos reconstruídos:\n", X_recon[:10])
```

### 39. Convolução 2D (im2col)  _(difícil)_

Implemente convolução 2D 'valid' de uma imagem `I` por um kernel `K` usando `sliding_window_view` para gerar blocos e multiplicação matricial.

```python
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view

def conv2d_valid(I, K):
    I = np.asarray(I)
    K = np.asarray(K)

    kH, kW = K.shape
    H, W = I.shape

    windows = sliding_window_view(I, (kH, kW))
    out_H, out_W = windows.shape[:2]

    windows_2d = windows.reshape(-1, kH * kW)
    K_flat = K.ravel()

    out_flat = windows_2d @ K_flat
    out = out_flat.reshape(out_H, out_W)

    return out

I = np.arange(25).reshape(5, 5)
K = np.array([[1, 0],
              [0, -1]])

Y = conv2d_valid(I, K)
print(Y)
```

### 40. Sistemas lineares e resíduo  _(difícil)_

Gere uma matriz A (100×100) bem-condicionada, um vetor b e resolva `Ax=b` com `np.linalg.solve`. Compute o resíduo relativo `||Ax−b||/||b||` e `cond(A)`.

```python
import numpy as np

rng = np.random.default_rng(42)

A = np.eye(100) + 0.01 * rng.normal(size=(100, 100))
b = rng.normal(size=100)

x = np.linalg.solve(A, b)

r = A @ x - b
rel_res = np.linalg.norm(r) / np.linalg.norm(b)
condA = np.linalg.cond(A)

print("Resíduo relativo:", rel_res)
print("cond(A):", condA)
```

### 41. Regressão linear (OLS) com NumPy  _(difícil)_

Dado `X` (N×d) e `y`, calcule os coeficientes OLS usando `np.linalg.lstsq` e avalie RMSE no treino.

```python
import numpy as np

rng = np.random.default_rng(42)
N, d = 100, 3
X = rng.normal(size=(N, d))
true_beta = np.array([1.5, -2.0, 0.7])
y = X @ true_beta + rng.normal(scale=0.5, size=N)

X_design = np.c_[np.ones(len(X)), X]

beta_hat, *_ = np.linalg.lstsq(X_design, y, rcond=None)

y_pred = X_design @ beta_hat
rmse = np.sqrt(np.mean((y_pred - y) ** 2))

print("Coeficientes (intercepto + betas):", beta_hat)
print("RMSE de treino:", rmse)
```
